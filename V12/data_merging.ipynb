## Adding new attributes to CBC data

- New_Attribute_1 = Current Negative Trades / TotalNumberOfTrades
- New_Attribute_2 = Historically Negative Trades / TotalNumberOfTrades
- New_Attibute_3 = (Current Negative Trade + Historically Negative Trade) /  TotalNumberOfTrades
- New_Attribute_4 = CurrentPastDue / TotalNumberOfTrades

```python
data = pd.read_excel("/content/drive/MyDrive/LCF_Internal/Renewable_Model/EDA Approach-1/XGBoost Implementation/v12/merged_csvs/ren_cbc_merged.xlsx")
```

```python
data["current_negative_per_trade"] = data["CurrentNegativeTrades"]/data["TotalNumberOfTrades"]
```

```python
data["historical_negative_per_trade"] = data["HistoricalNegativeTrades"]/data["TotalNumberOfTrades"]
```

```python
data["curr_hist_negative_per_trade"] = (data["HistoricalNegativeTrades"] + data ["CurrentNegativeTrades"])/data["TotalNumberOfTrades"]
```

```python
data["current_past_due_per_trade"] = data["CurrentPastDue"]/data["TotalNumberOfTrades"]
```

```python
data["OpportunityID"] = data["OpportunityID_ren"]
data.drop("OpportunityID_ren", axis = 1, inplace = True)
```
