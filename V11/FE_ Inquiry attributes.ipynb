```python
def inquiry_attributes_FE(df):
  for i in ['inquirycount03m', 'inquirycount12m', 'inquirycreditcount03m',
       'inquirycreditcount12m', 'inquiryhighriskcount03m',
       'inquiryhighriskcount12m', 'inquiryothercount03m',
       'inquiryothercount12m']:
    # if i == 'OpportunityID' or i == 'WriteOff_YN':
    #   continue
    df[i+'_value'] = df[i].apply(lambda x: 0 if x < 0 else x)
    df[i+'_flag'] = df[i].apply(lambda x: 1 if x < 0 else 0)
    df.drop(i,axis = 'columns',inplace = True)
```

```python
def inquiry_attributes_FE_categorical(df):
  df[['inquiryconsumeraddress_f1','inquiryconsumeraddress_f2','inquiryconsumeraddress_f3',
     'inquiryconsumerphone_f1','inquiryconsumerphone_f2','inquiryconsumerphone_f3']] = 0
  for i in ['inquiryconsumeraddress','inquiryconsumerphone']:
    df[i+'_f1'] = df[i].apply(lambda x: 1 if x == -1 else 0)
    df[i+'_f2'] = df[i].apply(lambda x: 1 if x == 0 else 0)
    df[i+'_f3'] = df[i].apply(lambda x: 1 if x == 1 else 0)
    df.drop(i,axis = 'columns',inplace = True)
```

```python
inquiry_attributes_FE(df)
inquiry_attributes_FE_categorical(df)
df
```
