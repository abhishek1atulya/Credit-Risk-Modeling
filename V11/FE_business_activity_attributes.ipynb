```python
def business_activity_attributes_FE(df):
  for i in ['businessaddrcount','businessrecordtimeoldest','businessrecordtimenewest']:
    # if i == 'OpportunityID' or i == 'WriteOff_YN':
    #   continue
    df[i+'_value'] = df[i].apply(lambda x: 0 if x < 0 else x)
    df[i+'_flag'] = df[i].apply(lambda x: 1 if x < 0 else 0)
    df.drop(i,axis = 'columns',inplace = True)
```

```python
df = pd.get_dummies(df, columns=['businessrecordupdated12m','businessactivity06m','businessactivity03m','businessactivity12m'])
```
